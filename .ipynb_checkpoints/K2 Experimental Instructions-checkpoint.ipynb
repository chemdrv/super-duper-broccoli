### Quenching 

A quencher can be any molecule that provides an additional pathway for the excited state (A$^*$) to return to the ground state (A) without emitting a photon of fluorescence. Typical quenchers do not undergo permanent reaction with the excited state and are not consumed during reaction. Looking at the reaction scheme, it is clear that fluorescence intensity will decrease with the addition of quencher because an additional, competitive pathway for decay of the excited state is present when quencher is added. 

When examining your K2 decay traces, there are two types of quenching to look out for - dynamic and static quenching. Dynamic quenching, also known as collisional quenching, has been discussed previously in this notebook, but will be elaborated on now. 

A Stern-Volmer plot is used to plot concentration of quencher [Q] against τ$_o$/τ. The slope of this plot is k$_Q$*τ$_o$ which is also equal to k$_D$. 

Below is a figure (Figure 1$^6$) that shows the Stern-Volmer plots for dynamic (collisional) quenching and static quenching. Note that a linear Stern-Volmer plot does not differentiate between dynamic or static quenching since both will have linear plots. The difference in dynamic and static quenching will mainly come from lifetime measurements. 

Dynamic quenching will result in a decrease in lifetime over time, because quenching is an additional rate process that depopulates the excited state. 

Static quenching will result in no decrease in lifetime, but instead a decrease in intensity of the flouorescence. For the sake of our experiment, this decrease in intensity will be shown through Vmax, the maximum voltage recorded in a trace. 



Additionally, there are circumstances where there can be a combination of static and dynamic quenching. Figure 2 shows some sample Stern-Volmer plots that compare the variations of quenching types. 

Note that at higher quencher concentrations, a dynamic quencher will appear have combined quenching. This may be due to the fact that when high concentrations of quencher is present, a large amount of the quencher molecules are in close proximity to the fluorophore.$^7$

### Procedure

#### Prepare Ru-bipy Stock Solution

The Nd:YAG laser used in these experiments has a wavelength of 532 nm. At this wavelength, Ruthenium-tris(2,2’-bipyridyl) dichloride (Ru-bipy) has a molar absorptivity of ca. 786 M-1s-1.$^1$$^0$ Prepare a Ru-bipy solution having an optical density (OD) of ca. 0.1 to 0.3 by adding Ru-bipy to ca. 500 mL DI. Beer’s Law can be used to estimate the amount of Ru-bipy required to prepare a solution having an OD between 0.1 and 0.3 at 532 nm. Measure the absorbance with available spectrophotometer. Add additional Ru-bpy as necessary to achieve the desired OD.

#### Prepare Quencher Solutions

Pick two quenchers, one of negative charge and one of positive charge (e.g., potassium ferricyanide and iron (III) chloride). The charge of a quencher relative to the charge of a fluorophore affects the bimolecular quenching rate constant. This should be kept in mind when deciding concentrations of stock quencher solutions. Prepare 10 mL solutions of each quencher by adding a known amount of quencher to a 10 mL volumetric flask and diluting to 10 mL with DI. If using iron (III) chloride as a quencher, it should be noted that iron (III) chloride is highly hygroscopic and should be weighed out quickly.

#### Perform Quenching Experiments

Add exactly 80 mL of Ru-bipy stock solution to the apparatus provided, taking note of the initial volume reading. Oxygen is known to quench the excited state of Ru-bipy.$^1$$^1$ Degas the Ru-bpy solution for at least 10 minutes by bubbling N$_2$ gas through the solution. Once the Ru-bipy solution is adequately degassed, begin flowing the solution through to the sample stage. Using the provided oscilloscope, save the decay trace for the unquenched solution. Stop flowing the Ru-bipy solution. Take note of the volume reading and the filename associated with the volume change. 

Depending on the concentration of prepared quencher solution, add between 0.100 mL and 1.000 mL of first quencher solution to the apparatus using a microliter syringe. Wait ca. one minute to allow the quencher to diffuse completely through the Ru-bipy solution. Flow solution through to the sample stage. Obtain decay trace for this solution as described above. Stop flowing. Take note of the volume of quencher solution added, the volume reading on the apparatus and the filename associated with the volume change and quencher addition. Continue adding quencher incrementally until Ru-bipy fluorescence is ca. 90% quenched. Depending on whether the quenching process is static or dynamic, the extent of quenching can be monitored, respectively, as a decrease in maximum fluorescence intensity or decrease in fluorescence lifetime. Repeat process for second quencher.


#### Variations in Ionic Strength

Changes in ionic strength influence the quenching of fluorophores.$^1$$^2$ Using the remaining Ru-bipy stock solution, prepare two Ru-bipy solutions of different ionic strengths by adding an appropriate amount of KCl. Prepare one Ru-bipy solution having an ionic strength between 0.01 M and 0.10 M, and prepare a second Ru-bipy solution having an ionic strength of 0.20 M. Repeat the quenching experiments described above for one of the quenchers used with the Ru-bipy solutions prepared.
