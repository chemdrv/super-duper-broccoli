## Introduction

Each month more than 100 incidents of laser flashes being observed by aircraft pilots while in flight are reported to the NTSB. Photochromic systems are being made into rapid response optical filters to protect against such inadvertent laser blinding. To be a good optical filter, not only does the response of the filter to be have to be rapid, but it must also return to a clear state rapidly once the detected treat is no longer present.  This experiment seeks to study the photochromic response of Photorome I, shown below.

![Photorome I structure](photorome_I.png "Photorome I structure")

Transient absorption experiments have become an important tool to observe light driven chemical transformations and reactions. A brief description of transient absorption experimental systems can be obtained from numerous sources$^{1-3}$ This technique has become common with the introduction of inexpensive laser sources that produce intense pulses of light. One family of molecules that have received significant attention are the photochromic systems that change color upon excitation.$^{4-6}$ These systems can be described using a simple kinetic scheme shown below.

$$A + h\nu \rightarrow B$$
$$B \xrightarrow{k_T} A$$
$$\mathrm{Scheme\:I}$$

The first reaction occurs during the laser pulse to populate the product state, B. Once the pulse is complete, the product state B slowly reverst back to the starting material. This second process is thermally activated and can be very slow such that the color change lasts for days or longer (i.e. AB540p) or be very rapid so that the color change is so short it cannot be observed by eye. In this experiment, the color lasts only a few seconds. The rate of return will be dependent upon the relative stabilities of A and B (linear free-energy relationship$^7$) and of the viscosity of the solvent (Perrin relationship$^8$).

Linear free-energy relationships allow the correlation of the stabilities of the reactant and product states with the activation enthalpy.$^7$ In general, the more stable the product (relative to the starting material) the lower the activation enthalpy and the faster the reaction rate. Given that polar solvents stabilize polar molecules more than nonpolar molecules, the relative stabilities of reactants and products can be manipulated in solvents of differing polarity. Thus, the relative polarities of the reactant and product states can be determined through observing the rate of interconversion as a function of the dielectric constant of the solvent. In this lab, the relative polarities of the reactant and product will be used to change the free energy difference, and therefore the activation enthalpy and further the rate of the interconversion.

Molecular rotation, much like diffusion, is dependent upon solvent viscosity. More viscous solvents will slow the process according to the Perrin relationship for rotational diffusion.

$$1/\tau = 1/\tau_0 + \frac{\alpha k_B}{V} \frac{T}{\eta}$$ $% figure out how to label this as Equation 1$

In equation 1, the $\tau_0$ is the intrinsic rotational time constant$% reword this$, $k_B$ is the Boltzmann constant, $T$ is the temperature, $V$ is the probe volume, $\eta$ is the dynamic viscosity, and $\alpha$ is a correction term dependent upon the boundary conditions. Thus, a plot of $1/k$ versus $T/\eta$ should be linear, and the magnitude of the slope should describe the degree to which the size and shape of the probe changes during the transformation.

## Data Analysis: Model Fitting Example

Below is an example of using python to create a model fit of a transient absorption graph. The voltage is changed into $\Delta$OD through the equation $A$ = log($I_{0}$/$I$). $I$ is intensity, and it is represented by the voltage on the y-axis when you take the transient absorption.

```python
#import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#import the file and plot it to make sure that you have the right data
raw_time, raw_voltage = np.loadtxt("K5_Hexane_550nm.CSV", delimiter=',', unpack = True, skiprows=17, usecols=(3,4))
plt.plot(raw_time, raw_voltage)
```

```python
#calculate the average of the baseline to use as Io, then use the voltage as I and convert to delta OD. You should see a change in the y-axis values
initialIndex = 243
initialAverage = np.average(raw_voltage[:initialIndex])
print(initialAverage)
raw_deltaOD = np.log10(initialAverage/raw_voltage)
plt.plot(raw_time,raw_deltaOD)
```

```python
#cut the trace so that only the decay portion shows

print("Length of time array:", np.shape(raw_time)) #tells me how many x value points I have
index = 245
plt.plot(raw_time[index:], raw_deltaOD[index:]) #cuts off beginning of the plot to just show the decay, as shown by the : being put after the variable
time = raw_time[index:] #time of only the decay portion
deltaOD = raw_deltaOD[index:] 
```

```python
#this uses the first order equation to make a model fit. you will need k, your rate constant, to compare to the other rate constants in your analysis

def exponential(x, a, k, b):
    return a*np.exp(k*x) + b

optData_first, covData_first = curve_fit(exponential, time, deltaOD, p0=(1400,-.1,0)) 
print(f"a = {optData_first[0]}, k = {optData_first[1]}, b = {optData_first[2]}")
```

```python
#plots the raw data and the model fit on top of each other

plt.plot(time, deltaOD, color='slateblue') #plot of raw data
plt.plot(time, exponential(time, optData_first[0],optData_first[1], optData_first[2]), color='orange') #plot of line of best fit
plt.legend(("Raw data", "Model"))
plt.ylabel("$\Delta$OD")
plt.xlabel("Time (s)")
plt.subplots_adjust(hspace=0)
print("This is Hexane at 550nm")
```
