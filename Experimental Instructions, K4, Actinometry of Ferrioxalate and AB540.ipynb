## Introduction
There is sufficient solar energy striking the earth’s surface to maintain the energy consumption of all of its inhabitants. Indeed, the total energy from all sources consumed today is less than 1% of the solar energy available. All that is required is an efficient method to collect and store solar energy. The efficiency of light driven reactions is measured using a value called quantum yield. [<sup>1</sup>](#fn1) The quantum yield of a photochemical reaction is defined as the number of moles of product formed relative to the number of moles of photons absorbed:
$$
\begin{equation}
    \phi = \frac{\text{Moles of product formed}}{\text{moles of photons absorbed}}
\end{equation}
\tag{1}
$$

Two pieces of information are required before a quantum yield can be determined.  First, the number of photons absorbed must be known. Second, the number of moles of product must be measured.  These two quantities will be determined in order to obtain a quantum yield for the photochemical conversion.

During this experiment you will observe two types of chemical change: a carbon-carbon bond formation and a redox reaction that are driven by light. A consideration for photochemical reactions is the efficiency through which they consume energy. It is important to understand that light (a photon) can also be used as a reagent in chemical reactions. For a thermal reaction, for example: , stoichiometry is straightforward and well-defined. One molecule of A reacts with one molecule of B to give one molecule of product P.  In a photochemical reaction, for example:  the stoichiometry is not as easily defined. While it is true that each photon will lead to an excited state (1:1 stoichiometry), each excited state does not necessarily lead to product formation. Several processes are in competition making the fate of the excited state somewhat uncertain. [<sup>1</sup>](#fn1)
For example, an excited state may react to form products, it may fluorescence or phosphorescence, or it may convert the photon energy into heat (internal conversion) and thereby return non-radiatively to the ground state.  The net result is that not all absorbed photons lead to product formation and the quantum yield can be less than 1.  
    
In order to determine the number of photons absorbed, the photon flux of the light source must be determined. The units of flux are moles of photons per second (also called Einsteins per second). With the photon flux known and the duration of the illumination measured, the number of photons that have entered the sample can be determined.  If the sample has a high optical density at the wavelength of interest, all of the photons that entered the sample will be absorbed.  Thus, the number of photons absorbed is simply determined by measuring the exposure time.  Details about how to carry out quantum yield calculations are described by Parker and by Calvert and Pitts cited in the References section. [<sup>2-4</sup>](#fn2) It is strongly suggested that these references be consulted before coming to the lab.

Ferrioxalate solutions will be used to experimentally determine the photon flux.  The brilliant green ferrioxalate (ferri = Fe<sup>3+</sup>) solution is irradiated with UV light of a selected wavelength (360 nm will be used in this experiment).  A photochemical reaction results in reduction of Fe<sup>3+</sup> to Fe<sup>2+</sup>.  The yield of Fe<sup>2+</sup> can be measured by observing the intense red complex between Fe<sup>2+</sup> and 1,10-phenanthroline; a familiar reaction you may remember from Chem 300. This photochemical reaction has been thoroughly studied since the mid-1950s, and its quantum yield for product formation is well known. [<sup>2-4</sup>](#fn4) By measuring the yield of products of this actinometer reaction as a function of time, the photon flux of the illumination source can be determined using:
$$
\begin{equation}
    F=\frac{n_b}{\phi_b t(1-10^{-\epsilon[C]{l}})}
\end{equation}
\tag{2}
$$

In Equation 2, $\phi_b$ is the quantum yield for the actinometer (i.e. $\phi_b$ = 1.21 at 360 nm).[<sup>1-3</sup>](#fn3) 

Given Beer’s Law, we know that:
$$
\begin{equation}
O.D. = {\epsilon}[C]{l}
\end{equation}
\tag{3}
$$

O.D. is the absorbance of the actinometer solution at 360 nm. $n_b$ is the number of moles of Fe<sup>2+</sup> produced, and t is the irradiation time in seconds.  

The result of Equation 2, F, is the photon flux in Einsteins per second.

The chemical reactant of interest in this experiment is E-α-(2,5-dimethyl-3-furylethylidene) isopropylidene succinic anhydride (**a**), also known as aberchrome 540 or AB540.  When exposed to light, AB540 gets converted to its spiro-derivative, 7,7a-dihydro-2,4,7,7,7a-pentamethylbenzo[b]furan-5,6-dicarboxylic anhydride (**b**), also known as AB540p. Fulgides, such as AB540, change color upon exposure to light.[<sup>5</sup>](#fn5)  It is possible to monitor the color change of a known concentration of AB540 to measure both product formation and starting material depletion.  Notice that while the stoichiometry of the photon reactant is not certain, one mole of AB540 will give exactly one mole of AB540p (1:1 stoichiometry).  The number of moles of photons required to carry out this change, however, will not be 1:1.
