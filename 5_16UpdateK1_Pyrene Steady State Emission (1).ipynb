## Introduction
#### Background

Understanding how excited states are produced and the processes through which these states
decay is of fundamental importance. Production of excited states and observing emission from
these states are two important aspects of efficient OLED design. The vast majority of biological
imaging techniques use fluorescence to observe chemical charges within living tissue.

Differences in fluorescence intensity and wavelength relays information of the chemical
environment in which the fluorophore resides. The specific type of kinetic analysis techniques
used in this experiment is called Stern-Volmer fluorescence quenching.

The process of emission of a photon from an excited state that has the same spin as the ground state is called
fluorescence. Of course, fluorescence is not the only possible fate of an excited state. Other
competing processes include heat evolution (internal conversion), conversion to a triplet state
(inter-system crossing), or chemical reaction. Excited states can also react with other molecules
found in solution, quenchers, including dissolved oxygen and other specifically added
molecules.

A mechanistic scheme for the creation and decay of an excited state of molecule A
is shown below.

#### Scheme 1

$A+ hv \rightarrow\ A^*$

$A^* \overrightarrow{{k_{IC}}}\  A + \Delta $ 

$A^* \overrightarrow{{k_{ISC}}}\ A^3$

$A^* \overrightarrow{{k_F}}\ A + hν$

$A^* + Q \overrightarrow{{k_Q}}\ A + Q^* $


#### What It Means
During steady state excitation, for example in a fluorescence spectrometer, the rate of formation
of excited state ($A^*$) in this scheme is dependent upon the photon flux of the excitation source at
the wavelength of excitation ($I_{Ex}$), the extinction coefficient (ε) of the absorber A at the
wavelength of excitation, instrumental parameters such as the pathlength (l) of the cell in which
the sample is placed, and the concentration ([A]) of absorber in solution.
An expression for the flux of photons produced by the second to last step in Scheme 1, the
emission of fluorescence from the excited state, can be derived directly from the rate law. In this
case, the flux of photons will depend on the rate of generation of the excited state, which can be
written as $I_{Ex}$ε[A]l. The fluorescence flux will depend inversely on the rate constants of all of the
processes by which the molecule can be removed from the excited state. 

The overall expression
can be written as: Where all equations will be

$F_F=(\frac{k_FI_{Ex}ε[A]l}{k_F+k_{IC}+k_{ISC}+k_R+k_q[Q]})$ (1)

The denominator in equation 1 is the sum of rate constants for the excited state decay. The
inverse of this sum is known as the fluorescent lifetime of the excited state:

$τ_F = (\frac{1}{k_F+k_{IC}+k_{ISC}+k_R+k_q[Q]})$ (2)


The quantum yield of fluorescence is defined as the ratio of emitted photons to the photons
absorbed and is given by the product of the lifetime and the fluorescence rate constant:

$Φ_F = (\frac{k_F}{k_F+k_{IC}+k_{ISC}+k_R+k_q[Q]}) = τ_Fk_F$ (3)

#### Why It Is Important
In general, the quantum yield for any process is given by the product of the rate constant for that
process and the lifetime of the reactive species. The aim of this experiment is to determine the fluorescent lifetime of an excited singlet state.

Note that the measurement of the emission flux alone does not allow direct determination of the
lifetime even if l and ε are known because the fluorescence rate constant, $k_F$, is unknown. There
are ways to estimate $k_F$ from ground state absorption spectra, but these methods are generally not
accurate enough to be useful. Instead, we will use diffusional quenching and fluorescence
measurements with Stern-Volmer analysis to determine the lifetime of the excited state.

#### Quenching
A quencher can be any molecule that provides an additional pathway for the excited state ($A^*$) to
return to the ground state (A) without emitting a photon of fluorescence. Typical quencher do not
undergo permanent reaction with the excited state and are not consumed during reaction.

There are different types of quenching, such as dynamic and static quenching. Dynamic quenching is when the quencher collides with the moleucule of interest, in this case pyrene. Static quenching can be considerd like a chemical reaction, where the quencher reacts chemically with pyrene. In this experiment, only dyanmic queching is observed because the reaction is not time dependent. 

Looking at the reaction scheme, and reviewing equation 1, it is clear that the fluorescence
intensity will decrease with the addition of quencher because an additional, competitive pathway
for decay of the excited state is present when quencher is added.

#### How To Use It
Taking the ratio of the fluorescence fluxes, or quantum yields, with and without quencher, we
can write:

$ (\frac{F_{0}}{F_{Q}}) = (\frac{Φ_{F0}}{Φ_{FQ}}) = (\frac{k_F+k_{IC}+k_{ISC}+k_R+k_q[Q]}{k_F+k_{IC}+k_{ISC}+k_R}) = k_Qτ_{F0}[Q]+1$ (4)

Equation 4 is known as the Stern-Volmer Equation.1-5 The subscript “0” indicates there are no
added quenchers ( i.e. [Q] = 0). A plot of $ (\frac{F_{0}}{F_{Q}})$ versus [Q] should be linear with a slope given
by $k_Qτ$ and an intercept of 1. If $k_Q$ is known, or can be estimated, it is then possible to estimate
the lifetime, $τ_0$.

#### Secondary Important Information
##### Diffusion
Einstein and Smoluchowski developed a theory of molecular diffusion in non-interacting
solvents.
A full description of this theory is given in the reference material provided. The
relationship is given by:

$k_D = \frac{4πN_A(D_A+D_B)(r_A+r_B)p}{1000}$ (5)

Where $N_A$ is Avogadro’s number, $D_i$ and $r_i$ are the diffusion coefficient and interaction radius of
particle i, and p is the probability that the two molecules will react once they are together. The
diffusion coefficients ($D_i$) are given by:

$D_i= \frac{kT}{6π\eta r_i'}$ (6)

In this equation, η is the solvent viscosity and $r_i’$ is the hydrodynamic radius of particle i. It is
customary to assume that the reactants are spheres and that the hydrodynamic radius and
interaction radius are the same. If it is assumed that the diffusing molecules are of similar sizes,

Equation 4 can be reduced to:

$k_D = \frac{8RTp}{3η}$ (7)

Reactions where p=1 are known as diffusion-controlled reactions,3-6 indicating that the rate of
reaction depends only on how quickly the two reactant molecules can diffuse together and as
soon as they are together they undergo reaction (in other words, diffusion and not chemical
reaction is the rate-limiting step). Values of $k_D$ may be determined using Equation 7 and $k_D = k_Q$
as long as the reaction in question is diffusion-controlled. To determine whether a reaction is
diffusion-controlled, one could monitor the reactions of several different quenchers to see if they
all have similar reaction rates. Only the fasted quencher(s) can be diffusion controlled and it
(they) should have rates constants given by equation 7. The calculation of this rate constant, and
information about units, can be found in Turro $^2$
